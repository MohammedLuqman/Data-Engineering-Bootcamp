```python
from database import engine
import pandas as pd

query = " SELECT * FROM raw_quotes"
df = pd.read_sql(query, engine)

df.head()
```

```python
df.isna().sum()
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df['quote'] = df['quote'].str.strip()
df['author'] = df['author'].str.strip()
df['tags'] = df['tags'].str.strip()
```

```python
df.to_sql('processed_quotes', con=engine, if_exists='replace', index=False)
print("It's Done")
```
