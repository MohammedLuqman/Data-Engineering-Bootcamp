### Task 1:
**Research Multiple Inheritance in Python and explain it in simple words with a small example**

Multiple inheritance means a class can get features (methods and variables) from more than one parent class.


```python
class Father:
    def speak(self):
        print("I am the father")
        
class Mother:
    def talk(self):
        print("I am the mother")
        
class child(Father,Mother):
    pass

c = child()
c.speak()
c.talk()


    
```
